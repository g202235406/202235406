```python
Y=totall_data[['target']]
Y.head()
```

```python
X=totall_data.drop(['target'], axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=0.2,random_state=5)
```

```python
print("X_train.shape :", X_train.shape)

print("y_train.shape :", y_train.shape)

print("X_test.shape:", X_test.shape)

print("y_test.shape:", y_test.shape)
```

```python
clf_svm = SVC(random_state=0)
clf_svm.fit(X_train, y_train)

pred_svm = clf_svm.predict(X_test)

print("\n--- SVM Classifier ---")
cm_svm = confusion_matrix(y_test, pred_svm)
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)
cm_lr = confusion_matrix(y_test, pred_lr)
print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
cm_dt = confusion_matrix(y_test, pred_dt)
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Radom Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
cm_rf = confusion_matrix(y_test, pred)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```

```python
clf_nn = MLPClassifier(random_state=0)
clf_nn.fit(X_train, y_train)

pred_nn = clf_nn.predict(X_test)

print ("\n--- Neural Network Classifier ---")
cm_nn = confusion_matrix(y_test, pred_nn)
print (accuracy_score(y_test, pred_nn))
print (confusion_matrix(y_test, pred_nn))
```

```python
model = KNeighborsClassifier(n_neighbors = 4)
model.fit(X_train, y_train.values.ravel())
```

```python
model.score(X_test, y_test.values.ravel())
```

```python
a = model.score(X_test, y_test)
a
```

```python
for n in range(3, 30):
    model = KNeighborsClassifier(n_neighbors = n)
    model.fit(X_train, y_train.values.ravel())
    score = model.score(X_test, y_test)
    if score > a:
        print(n)
        a = score
        max = n
model.n_neighbors = max
model.fit(X_train, y_train.values.ravel())
max_S = model.score(X_test, y_test)
print("정확도가 가장 높은 최근접 이웃 개수는 정확도가 %.9f%%인 %d 입니다."%(max_S*100, max))
```
